### Construct CaOH ${}^2\Sigma^+$ states  in $|F, m\rangle$ basis

```julia
# Enumerate all states of N = 1 in a Hund's case (b) basis
η = (S = 1//2, I = 1//2, Λ = 0, N = 1, J = 1//2, F = 1, M = 0)
QN_bounds = (N = (1, 1),)
basis = enumerate_states(η, HundsCaseB, QN_bounds)
```

```julia
states = states_from_basis(basis)
;
```

```julia
const gS = 2.0023
const e = 1.60217662e-19
const me = 9.10938356e-31
const μB = (e * ħ / (2me)) * (1 / h) * 1e-4 * 1e-6 # 1e-4 = Tesla to Gauss, 1e-6 = Hz to MHz
;
```

```julia
# Spectroscopic constants
# From Scurlock et al.
const BX = @with_unit 10023.0841 "MHz"
const DX = @with_unit 1.154e-2 "MHz"
const γX = @with_unit 34.7593 "MHz"
const bFX = @with_unit 2.602 "MHz"
const cX = @with_unit 2.053 "MHz"
const μX = 1e-6 * 1.465 * 10^2 * (1e-21 / c) / h # 1e-6 = Hz to MHz
;
```

```julia
# Define the zero-field Hamiltonian operator
H_operator = 
    BX * Rotation + 
    DX * RotationDistortion + 
    γX * SpinRotation + 
    bFX * Hyperfine_IS + 
    (cX/3) * Hyperfine_Dipolar
;
```

```julia
H = Hamiltonian(basis=basis, H_operator=H_operator)
;
```

```julia
# Add the Stark operator to the zero-field Hamiltonian operator 
HStark(E) = H_operator + μX * E * Stark
H = Hamiltonian(basis=basis, H_operator=HStark);

Es = 0:5000:100000
eigenstates = zeros(length(Es), length(H.basis), length(H.basis))
energies = zeros(length(Es), length(H.basis))
@time begin
    for (i, E) in enumerate(Es)
        es, vs = solve(H, E=E)
        eigenstates[i, :, :] = vs
        energies[i, :] = es
    end
end
plot(Es, energies, legend=nothing)
```

```julia
η = (S = 1//2, I = 1//2, Λ = 0, N = 1, J = 1//2, F = 1, M = 0)
QN_bounds = (N = (1, 1),)
basis_N1 = enumerate_states(η, HundsCaseB, QN_bounds)

# Add the Stark operator to the zero-field Hamiltonian operator 
HZeeman(B) = H_operator + gS * μB * B * Zeeman
H = Hamiltonian(basis=basis_N1, H_operator=HZeeman);

Bs = 0:0.01:1
eigenstates = zeros(length(Bs), length(H.basis), length(H.basis))
energies = zeros(length(Bs), length(H.basis))
@time begin
    for (i, B) in enumerate(Bs)
        es, vs = solve(H, B=B)
        eigenstates[i, :, :] = vs
        energies[i, :] = es
    end
end
plot(Bs, energies, legend=nothing)
```

### Construct CaOH ${}^2\Pi_{1/2}$ states  in $|F, m\rangle$ basis

```julia
# Enumerate all states of Λ = -1 in a Hund's case (a) basis
η = (I = 1//2, S = 1//2, Λ = -1, J = 3//2, Ω = -1//2, Σ = 1//2, F = 1, M = 1)
QN_bounds = (J = [1//2, 3//2],)
basis_Λm1 = enumerate_states(η, HundsCaseA, QN_bounds)
```

```julia
# Enumerate all states of Λ = +1 in a Hund's case (a) basis
η = (I = 1//2, S = 1//2, Λ = +1, J = 3//2, Ω = 1//2, Σ = 1//2, F = 1, M = 1)
QN_bounds = (J = [1//2, 3//2],)
basis_Λp1 = enumerate_states(η, HundsCaseA, QN_bounds)
```

```julia
basis = [basis_Λm1; basis_Λp1]
states = states_from_basis(basis);
```

```julia
# Spectroscopic constants for the A state
# From Li & Coxon (1995)
const BA = 0.3412200 * c * 1e2
const AA = 66.8181 * c * 1e2
const pA = -0.04287 * c * 1e2
const qA = -0.3257e-3 * c * 1e2
const aA = 1.0 * 1e4
const bFA = 1.0 * 1e4
const cA = 1.0 * 1e4
const dA = 1.0 * 1e4
;
```

```julia
# Define the zero-field Hamiltonian operator
H_caseA_operator = 
    BA * Rotation +
    AA * SpinOrbit + 
    BA * SpinUncoupling + 
    qA * ΛDoubling_q +
    (pA + 2qA) * ΛDoubling_p2q +
    aA * Hyperfine_IL +
    bFA * Hyperfine_IF +
    cA * Hyperfine_Dipolar_c +
    dA * Hyperfine_Dipolar_d
;
```

```julia
H_caseA = Hamiltonian(basis=basis, H_operator=H_caseA_operator);
```

```julia
es, vs = solve(H_caseA);
```

```julia
coeffs = vs[1,:,1]
x = State(basis, coeffs)
```

### Construct the $X{}^2\Sigma^+(01^10)$ states in a Hund's case (b) basis

```julia
# Spectroscopic constants for X(010)
# From Table III, Li and Coxon (1995)
# Units are MHz
const Tv = 352.9259 * c * 1e-4 # the fundamental bending frequency
const Bv = 0.333458 * c * 1e-4
const Dv = 0.3930e-6 * c * 1e-4
const γv = 0.001184 * c * 1e-4
const qv = -0.7181e-3 * c * 1e-4
const bFX = 1.0 #2.602
const cX = 1.0 #2.053
;
```

```julia
η = (S = 1//2, I = 1//2, Λ = 1, N = 1, J = 1//2, F = 1, M = 0)
QN_bounds = (N = (1, 1),)
basis_p = enumerate_states(η, HundsCaseB, QN_bounds)
;
```

```julia
η = (S = 1//2, I = 1//2, Λ = -1, N = 1, J = 1//2, F = 1, M = 0)
QN_bounds = (N = (1, 1),)
basis_m = enumerate_states(η, HundsCaseB, QN_bounds)
;
```

```julia
# We order the basis according to increasing M_F quantum number, which ensures that M_F is a good quantum number 
# for the zero-field eigenstates
basis = [basis_m; basis_p]
ordered_basis = order(basis, :M)
;
```

```julia
# Define the zero-field Hamiltonian operator
H_operator = 
    Bv * Rotation + 
    Dv * RotationDistortion + 
    γv * SpinRotation +
    bFX * Hyperfine_IS + 
    (cX/3) * Hyperfine_Dipolar +
    qv * ℓDoubling
;
```

```julia
H_X010_bare = Hamiltonian(basis=ordered_basis, H_operator=H_operator)
H_Stark = Hamiltonian(basis=ordered_basis, H_operator=Stark)
H_Zeeman = Hamiltonian(basis=ordered_basis, H_operator=Zeeman)
H = zeros(ComplexF64, length(ordered_basis), length(ordered_basis));
;
```

```julia
# Define function to easily update the full Hamiltonian, saving the full Hamiltonian in H (by overwriting)
function H_X010_Field!(H, H0, H_Stark, H_Zeeman, B, E)
    @. H = H0 + μX * E * H_Stark + gS * μB * B * H_Zeeman
    return nothing
end;
```

```julia
B = 0.0
Es = 0:0.1:500
eigenstates      = zeros(length(Es), length(ordered_basis), length(ordered_basis))
energies         = zeros(length(Es), length(ordered_basis))
magnetic_moments = zeros(length(Es), length(ordered_basis))

@time begin
    for (i, E) in enumerate(Es)
        H_X010_Field!(H, H_X010_bare.M, H_Stark.M, H_Zeeman.M, B, E)
        es, vs = eigen(H)
        eigenstates[i, :, :] = real.(vs)
        energies[i, :] = real.(es)
        
        ΔB = 1e-6
        H_X010_Field!(H, H_X010_bare.M, H_Stark.M, H_Zeeman.M, B + ΔB, E)
        es′, vs′ = eigen(H)
        es′, vs′ = real.(es′), real.(vs′)

        magnetic_moments[i, :] = (es′ .- es) / (μB * ΔB)
    end
end
plot(Es, magnetic_moments, legend=nothing)
```

```julia
E = 0.0
Bs = 0:0.1:10
eigenstates      = zeros(length(Bs), length(ordered_basis), length(ordered_basis))
energies         = zeros(length(Bs), length(ordered_basis))

@time begin
    for (i, B) in enumerate(Bs)
        H_X010_Field!(H, H_X010_bare.M, H_Stark.M, H_Zeeman.M, B, E)
        es, vs = eigen(H)
        eigenstates[i, :, :] = real.(vs)
        energies[i, :] = real.(es)
    end
end
plot(Bs, energies, legend=nothing)
```
