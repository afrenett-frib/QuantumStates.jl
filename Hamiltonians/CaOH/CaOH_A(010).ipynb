```julia
using Revise
```

```julia
using QuantumStates
using UnitsToValue
```

```julia
const _μB = (μ_B / h) * (1e-6 * 1e-4)
const _gL = 1.0
const _gS = gS
;
```

```julia
H_operator = :(
    T * DiagonalOperator +
    B * Rotation +
    A * SpinOrbit + 
    q_e * ΛDoubling_q +
    p_e * ΛDoubling_p2q + q_e * (2ΛDoubling_p2q) +
    q_v * (ℓDoubling) + 
    ϵω2 * (RennerTeller/2) #+
    # b00_A * (Hyperfine_IF - Hyperfine_IL)
);
```

```julia
# Spectroscopic constants for the A(010) state
# Taken from Li & Coxon (1995)
parameters = @params begin
    T = 16360.389 * 299792458 * 1e-4
    B = 0.3402609 * 299792458 * 1e-4
    A = 67.0951 * 299792458 * 1e-4
    q_e = -0.3072e-3 * 299792458 * 1e-4
    p_e = -0.04472 * 299792458 * 1e-4
    q_v = -0.6978e-3 * 299792458 * 1e-4
    ϵω2 = -36.2634 * 299792458 * 1e-4
    # b00_A = 0.
end;
```

```julia
QN_bounds = (
    v_1 = 0,
    v_2 = 1,
    ℓ = (-1,1),
    v_3 = 0,
    Λ = (-1,1),
    I = 1/2,
    S = 1/2,
    J = 1/2:5/2
    )
basis = enumerate_states(HundsCaseA_LinearMolecule, QN_bounds)
;
```

```julia
CaOH_A010 = @time Hamiltonian(basis=basis, operator=H_operator, parameters=parameters)
@time evaluate!(CaOH_A010)
@time solve!(CaOH_A010)
;
```
